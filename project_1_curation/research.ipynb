```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Idea
- Research the returns on investment for the norwegian oil fund. Do they beat the market? Are they strong investors? some questions like that.

Data is found at: https://www.nbim.no/no/oljefondet/avkastning/

Datasets:

returns_per_category.xls: The returns NBIM have had for the different categories.

Aggregate_returns.xls: The aggregate returns each year.

Holding_industry.xlsx: The market value, ownership, and industry of different companies.

```python
returns = pd.read_excel('data/return_per_category.xls')
```

#### Possible questions we will seek to answer:

- Where does Norway choose to invest our money?

- Do they beat the market with their investment choices?

```python
fig, ax = plt.subplots(2,2, figsize=(10,7))

years = returns.iloc[:,0]
ax[0,0].bar(years, returns.iloc[:,1], color = 'green')
ax[0,0].set_title('Stocks')

ax[0,1].bar(years, returns.iloc[:,2], color = 'orange')
ax[0,1].set_title('Rentepapir')

ax[1,0].bar(years, returns.iloc[:,3], color = 'blue')
ax[1,0].set_title('eindom')

ax[1,1].bar(years, returns.iloc[:,4], color = 'orange')
ax[1,1].set_title('fornybar')
```

```python
plt.plot(years, returns.iloc[:,1], label='Stocks')
plt.plot(years, returns.iloc[:,2], label='Rentepapir')

plt.axhline(0,0,1, color='black')
plt.legend()
plt.show()
```

Can disregard eindom and renewable energy from our research as it has not been invested in for long and it only makes up a combined 9% of NBIM's investment. The stocks make up about 70% of the investment as of september 2024. 

From here we may raise the hypothesis that the Rentepapir is a more secure investment as opposed to stocks and make a case for it making up more of NBIM's investments. Can do so via the use of one tailed hypothesis testing perhaps. Or we could take the path of comparing expected returns which will probably lead to the well known fact that stocks carry high risk but also high returns.
